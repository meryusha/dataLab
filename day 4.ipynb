```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', dodge = True)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips, hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips, hue = 'sex', markers = ['o', 'v'])
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips, hue = 'sex', scatter_kws = {'s':100})
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips, hue = 'sex', col = 'time', row = 'sex', size= 5, aspect = 0.9)
```
