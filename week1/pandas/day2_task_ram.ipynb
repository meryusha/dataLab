```python
import numpy as np

#creating array of 10 zeros
arr = np.zeros(10)
print(arr)

#create array of 10 ones
arr = np.ones(10)
print(arr)

#create array of 10 fives
arr = np.ones(10) * 5
print(arr)

#create array of numbers from 10 to 50 (including)
arr = np.arange(10, 51)
print(arr)

#create array 3x3 with numbers 0 to 8
arr = np.arange(9).reshape(3, 3)
print(arr)

#create 3x3 identity matrix
arr = np.eye(3)
print(arr)

#random number 0 -1
num = np.random.rand(1)
print (num)

#25 random numbers
arr = np.random.random((5,5)) + 1
print(arr)

#matrxi below
arr = np.linspace(0.01, 1, 100).reshape(10, 10)
print(arr)

#20 uniformly inreasing numbers from 0 to 1
arr = np.linspace(0, 1, 20)
print(arr)

#task given
mat = np.arange(1, 26).reshape(5,5)
print(mat[2:, 1:])

print(mat[3, 4])

#print(mat[0:3, 1])
print(np.hsplit(mat[0:3, :], 5)[1])
print(mat[:, 3:4])

print(mat[4, :])

print(mat[3:5, :])

print(np.sum(mat))
print(np.std(mat))
print(np.sum(mat, axis = 0))
```
