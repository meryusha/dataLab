```python
#137
def get_reason ( title):
    return title.split(':')[0]
dataset['reason'] = dataset['title'].apply(get_reason)
```

```python
dataset['timeStamp'] = pd.to_datetime(dataset['timeStamp'])
#a = pd.to_datetime(dataset['timeStamp'])
```

```python
def set_hour(date):
    return date.hour
def set_month(date):
    return date.month
def set_day(date):
    return date.day_name()

dataset['Hour'] = dataset['timeStamp'].apply(set_hour)
dataset['Month'] = dataset['timeStamp'].apply(set_month)
dataset['Day of Week'] = dataset['timeStamp'].apply(set_day)
```

```python
#dataset
```

```python
dataset.drop('timeStapm', axis = 1, inplace = True)
```

```python
dataset
```

```python
sns.countplot(x = 'Day of Week', data = dataset, hue = 'reason')
plt.legend(bbox_to_anchor= (1.05, 1), loc =2, borderaxespad = 0)
```

```python
sns.countplot(x = 'Month', data = dataset, hue = 'reason')
plt.legend(bbox_to_anchor= (1.05, 1), loc =2, borderaxespad = 0)
```

```python
def to_date(date):
    return date.date()

dataset['date'] = dataset['timeStamp'].apply(to_date)
```

```python
d = dataset.groupby('date').count()
#d
d['twp'].plot()

plt.tight_layout()
```

```python
#d = dataset[dataset['reason'] == 'Traffic'].groupby('date').count()
#d
def plot_graph_reason(res, number_graphs, i):
    d = dataset[dataset['reason'] == res].groupby('date').count()
    plt.subplot(number_graphs, 1, i )
    plt.title(res)
    plt.ylim(0, 600)
    plt.plot(d['twp'])
    
number_graphs = dataset['reason'].unique()
print(number_graphs)
i = 1
plt.subplots(len(number_graphs),1,figsize=(10,10))
for res in number_graphs:
    plot_graph_reason(res, len(number_graphs), i)
    i += 1

plt.subplots_adjust(hspace=0.1, wspace=0.01)

plt.tight_layout()
#d = dataset[dataset['reason'] == 'Traffic'].groupby('date').count()
#plt.plot(d['twp'])
```
