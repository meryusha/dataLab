```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math

x = np.linspace (0, math.pi*2,50)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r')
plt.xlabel('X axis title')
plt.ylabel('Y axis lable')
plt.title('String title')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import math
x = np.arange(0, 101)
fig2 = plt.figure()
ax1 = fig2.add_axes([0, 0, 1, 1])
ax2 = fig2.add_axes([0.2, 0.5, 0.2, 0.2])
ax1.plot(x, x*2)

ax2.plot(x, x*2)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

#plt.title('title')
```

```python

x = np.arange(0, 101)
fig2 = plt.figure()
ax1 = fig2.add_axes([0, 0, 1, 1])
ax2 = fig2.add_axes([0.2, 0.5, 0.4, 0.4])
ax1.plot(x, x**2)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

x2 = np.arange(20, 23)
ax2.plot(x2, x2*2)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_ylim(30, 50)
ax2.set_title('zoom')
#plt.title('title')
```

```python
x = np.arange(0, 101)
plt.subplots_adjust(hspace=0.4, wspace=0.4)
plt.subplot(1, 2, 1)
plt.plot(x, 2* x, c = 'blue', ls='--')

plt.subplot(1, 2, 2)
plt.plot(x, x**2, c = 'red', ls = '-')
plt.plot()
```
